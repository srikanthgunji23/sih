```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
from tensorflow.compat.v1.losses import sparse_softmax_cross_entropy
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text

def summarize_legal_case(document):
    model_name = "sshleifer/distilbart-cnn-12-6"
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)
    
    # Split the document into chunks of 2000 words
    chunk_size = 500
    chunks = [document[i:i + chunk_size] for i in range(0, len(document), chunk_size)]

    # Summarize each chunk and concatenate the summaries
    summaries = [summarizer(chunk, max_length=50, min_length=10, length_penalty=2.0, num_beams=4, early_stopping=True)[0]['summary_text'] for chunk in chunks]
    summary = ' '.join(summaries)

    return summary

def save_summary_to_txt(summary, output_path='1.txt'):
    with open(output_path, 'w') as file:
        file.write(summary)

def main():
    # Replace 'your_legal_case.pdf' with the path to your PDF file
    pdf_path = 'lol.pdf'

    # Extract text from the PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    # Summarize the extracted text
    summary = summarize_legal_case(pdf_text)

    # Save the summary to a text file
    save_summary_to_txt(summary)

if __name__ == "__main__":
    main()
```
